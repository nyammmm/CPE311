```python
wins = np.array(df['wins'])
losses = np.array(df['losses'])
score = np.array(df['Score'])
avewin = int(np.average(wins))
aveloss = int(np.average(losses))
avescore = int(np.average(score))
print("Average team wins: " + str(avewin))
print("Average team losses: " + str(aveloss))
print("Average team score: " + str(avescore))
```

```python
wins = np.array(df['wins'])
losses = np.array(df['losses'])
score = np.array(df['Score'])
avewin = int(np.median(wins))
aveloss = int(np.median(losses))
avescore = int(np.median(score))
print("Average team wins: " + str(avewin))
print("Average team losses: " + str(aveloss))
print("Average team score: " + str(avescore))
```
