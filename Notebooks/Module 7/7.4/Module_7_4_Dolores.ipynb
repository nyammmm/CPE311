## **SETUP**

```python
import pandas as pd
long_df = pd.read_csv('/content/long_data.csv', usecols=['date', 'datatype', 'value']).rename(columns={'value' : 'temp_C'}).assign(date=lambda x: pd.to_datetime(x.date), temp_F=lambda x: (x.temp_C * 9/5) + 32)
long_df.head()
```

## **TRANPOSING**

```python
long_df.head().T
```

## **PIVOTING**

```python
pivoted_df = long_df.pivot(index='date', columns='datatype', values='temp_C')
pivoted_df.head()
```

```python
pd.pivot(long_df, index='date', columns='datatype', values='temp_C').head()
```

```python
pivoted_df.describe()
```

```python
pivoted_df = long_df.pivot(index='date', columns='datatype', values=['temp_C', 'temp_F'])
pivoted_df.head()
```

```python
pivoted_df['temp_F']['TMIN'].head()
```

```python
multi_index_df = long_df.set_index(['date', 'datatype'])
multi_index_df.index
```

## **MELTING**

```python
wide_df = pd.read_csv('/content/wide_data.csv')
wide_df.head()
```

```python
melted_df = wide_df.melt(id_vars='date', value_vars=['TMAX', 'TMIN', 'TOBS'], value_name='temp_C', var_name='measurement')
melted_df.head()
```

```python
pd.melt(wide_df, id_vars='date', value_vars=['TMAX', 'TMIN', 'TOBS'], value_name='temp_C', var_name='measurement').head()
```

```python
wide_df.set_index('date', inplace=True)
wide_df.head()
```

```python
stacked_series = wide_df.stack()
stacked_series.head()
```

```python
stacked_df = stacked_series.to_frame('values')
stacked_df.head()
```
