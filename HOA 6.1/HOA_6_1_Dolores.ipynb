```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
print(salaries)
```

```python
#MEAN
def getmean(salaries):
  #GET THE SUM OF ALL NUMBERS IN ARRAY AND DIVIDE IT BY ITS SIZE
  mean = sum(salaries)/len(salaries)
  return mean

print(getmean(salaries))
```

```python
#MEDIAN
def getmedian(salaries):
  sortarr = sorted(salaries)
  midex = len(sortarr)//2
  if (len(sortarr) % 2 == 0):
    #IF THE SIZE OF ARRAY IS EVEN DIVED THE TWO MIDDLE NUMBERS
    median = (sortarr[midex-1]+sortarr[midex])/2
  else:
    #IF THE SIZE OF ARRAY IS ODD THEN THE MIDDLE IS THE MEDIAN
    median = sortarr[midex]
  return median

print(getmedian(salaries))
```

```python
#MODE
def getmode(salaries):
  #INITIALIZE A DICTIONARY TO STORE THE FREQUENCY
  frequendict = {}
  for i in salaries:
    #ITERATES TO GET THE FREQUENCY OF EACH NUMBER
    frequendict[i] = frequendict.get(i,0)+1
  maxc = max(frequendict.values())
  #FIND THE NUMBER WITH HIGHEST FREQUENCY
  mode = [i for i, count in frequendict.items() if count == maxc]
  freq = frequendict[mode[0]]
  return mode, freq

print(getmode(salaries))
```

```python
#SAMPLE VARIANCE
def getsamvar(salaries):
  index = len(salaries)
  #CALLING THE GETMEAN FUNCTION
  mean = getmean(salaries)
  #CALCULATE BY GETTING THE SUM OF SQUARED DIFFERENCE OF MEAN AND EACH SALARY
  variance = sum((i-mean)**2 for i in salaries)/(index-1)
  return variance

print(getsamvar(salaries))
```

```python
#STANDARD DEVIATION
def getstandev(salaries):
  #RAISE TO 1/2 TO GET THE STANDARD DEVIATION
  standev = getsamvar(salaries)**(1/2)
  return standev

print(getstandev(salaries))
```

**EXERCISE 2:**

```python
#RANGE
def getrange(salaries):
  range = max(salaries) - min(salaries)
  return range

print(getrange(salaries))
```

```python
#COEFFICIENT OF VARIATION & INTERQUARTAL RANGE
def getcviqr(salaries):
  mean = getmean(salaries)
  stdv = getstandev(salaries)
  cv = (stdv/mean)*100
  import statistics as st
  q1 = st.quantiles(salaries, n=4)[0]
  q3 = st.quantiles(salaries, n=4)[2]
  iqr = q3 - q1
  return cv, iqr

cv, iqr = getcviqr(salaries)
print("Coefficient of Variation: " + str(cv))
print("InterQuartal Range: " + str(iqr))
```

```python
#QUARTILE COEFFICIENT OF DISPERSION
def getqcd(salaries):
  import statistics as st
  q1 = st.quantiles(salaries, n=4)[0]
  q3 = st.quantiles(salaries, n=4)[2]
  iqr = q3 - q1
  median = getmedian(salaries)
  qcd = iqr/median
  return qcd

qcd = getqcd(salaries)
print("QCD: "+str(qcd))
```
