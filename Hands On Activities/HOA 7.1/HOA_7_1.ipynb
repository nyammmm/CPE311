### **Exercise 1**

```python
import pandas as pd
filepath1 = '/content/aapl.csv'
filepath2 = '/content/amzn.csv'
filepath3 = '/content/fb.csv'
filepath4 = '/content/goog.csv'
filepath5 = '/content/nflx.csv'

appl = pd.read_csv(filepath1)
amzn = pd.read_csv(filepath2)
fb = pd.read_csv(filepath3)
goog = pd.read_csv(filepath4)
nflx = pd.read_csv(filepath5)

dfs = [appl, amzn, fb, goog, nflx]
```

```python
appl['ticker'] = 'AAPL'
appl
```

```python
amzn['ticker'] = 'AMZN'
amzn
```

```python
fb['ticker'] = 'FB'
fb
```

```python
goog['ticker'] = 'GOOG'
goog
```

```python
nflx['ticker'] = 'NFLX'
nflx
```

```python
faang = pd.DataFrame()

for i in dfs:
  faang = faang.append(i, ignore_index=True)
```

```python
faang
```

```python
faang.to_csv('/content/faang.csv', index=False)
```

### **Exercise 2**

```python
faang['date'] = pd.to_datetime(faang['date'])
faang
```

```python
faang['volume'] = faang['volume'].astype(int)
```

```python
faang = faang.sort_values(by=['date', 'ticker'])
faang
```

```python
dfhighvlm = faang.nlargest(n=7, columns='volume')
dfhighvlm
```

```python
faangmelt = pd.melt(faang, id_vars=['date','ticker'])
faangmelt
```
